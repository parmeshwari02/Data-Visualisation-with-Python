```python


x = df['Team/NOC']
y1 = df['Gold']
y2 = df['Silver']
y3 = df['Bronze']

plt.figure(figsize=(50,40))
plt.xticks(fontsize=30)
plt.yticks(fontsize=30)

plt.xlabel('xlabel',fontsize=60)
plt.ylabel('ylabel',fontsize=60)

plt.bar(x, y1, color='y')
plt.bar(x, y2, bottom=y1, color='g')
plt.bar(x, y3, bottom=y1+y2, color='b')

plt.xticks(rotation=90)
plt.xlabel("Teams")
plt.ylabel("Medals")
plt.legend(["Gold", "Silver", "Bronze"],fontsize=60)
plt.title("Medals Scored by Teams", fontsize=60, fontweight='bold')
plt.show()
```

```python
plt.figure(figsize=(50,40))
sns.set(font_scale=3)
plt.xticks(rotation=90)
sns.scatterplot(x='Team/NOC',y='Gold',data=df)
sns.lineplot(x='Team/NOC',y='Silver',data=df)
sns.barplot(x='Team/NOC',y='Bronze',data=df)
```
